```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pandas_cat import pandas_cat
```

```python
weather = pd.read_csv('Dickerson_Weather_Station_Data_20240504.csv')
```

```python
weather['wasRaining'] = weather['RAIN ']>0
```

```python
x1 = weather

fig = plt.figure(figsize=(18,5))
ax1 = fig.add_subplot(1, 3, 1)
#plt.figure(figsize=(10, 6))
sns.countplot(data=x1, x='wasRaining', palette='viridis')
plt.xlabel('wasRaining')
plt.ylabel('Count individuals')
plt.title('weather - was raining?')
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

```python
def fun(t):
    if t <47 or t > 82:
        return 'less than 47F or more than 82F'
    else:
        return '47-82F'
```

```python
weather['Temperature'] = weather['TMP10-DK '].apply(fun)
```

```python
x1 = weather

fig = plt.figure(figsize=(18,5))
ax1 = fig.add_subplot(1, 3, 1)
#plt.figure(figsize=(10, 6))
sns.countplot(data=x1, x='Temperature', palette='viridis')
plt.xlabel('Temperature')
plt.ylabel('Count individuals')
plt.title('weather - Temperature')
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

```python
def funn(w):
    if w >6.3:
        return 'windy'
    else:
        return 'not-windy'
```

```python
weather['Wind'] = weather['SPD10-DK '].apply(funn)
```

```python
x1 = weather

fig = plt.figure(figsize=(18,5))
ax1 = fig.add_subplot(1, 3, 1)
#plt.figure(figsize=(10, 6))
sns.countplot(data=x1, x='Wind', palette='viridis')
plt.xlabel('Wind')
plt.ylabel('Count individuals')
plt.title('weather - Wind')
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```
